```python
import json
import random

test_examples = json.load(open('./processed_data/fire_test.json'))

# To generate random test example
def generate_random_test_example():
    random_example = [random.choice(test_examples)]
    test_path = './test.json'
    with open(test_path, 'w') as f:
      json.dump(random_example, f)

    return random_example
```
