```python
import pandas as pd
import matplotlib.pyplot as plt
import ast


file_path = './processed_data/log/fire/2025-03-06_15:06:13.415036/loss_avg_train.csv'
df = pd.read_csv(file_path, delimiter=";")


df['loss_avg'] = df['loss_avg'].apply(ast.literal_eval)

df['loss_avg'] = df['loss_avg'].apply(lambda x: x[0] if isinstance(x, list) else x)

plt.figure(figsize=(10, 5))
plt.plot(df['global_iteration'], df['loss_avg'], marker='o', linestyle='-')
plt.xlabel("Global Iteration")
plt.ylabel("Loss Average")
plt.title("Loss Average Over Iterations")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = "./processed_data/log/fire_eval_log/2025-03-06_22:02:36.950203/eval_test.csv"  # Change this if needed
df = pd.read_csv(file_path, sep=";")  # Assuming values are separated by semicolons


# NER metrics
ner_metrics = {
    "Micro Precision": df["ner_prec_micro"].values[0],
    "Micro Recall": df["ner_rec_micro"].values[0],
    "Micro F1-score": df["ner_f1_micro"].values[0],
    "Macro Precision": df["ner_prec_macro"].values[0],
    "Macro Recall": df["ner_rec_macro"].values[0],
    "Macro F1-score": df["ner_f1_macro"].values[0],
}

# Create bar plot for NER performance
plt.figure(figsize=(9, 6))
bars = plt.bar(ner_metrics.keys(), ner_metrics.values(), color=sns.color_palette("dark", n_colors=len(ner_metrics)), alpha=0.85)
plt.ylabel("Score (%)", fontsize=12)
plt.title("NER Performance", fontsize=14, fontweight="bold")
plt.xticks(rotation=45, fontsize=10)
plt.ylim(0, 100)
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Display values on bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f"{yval:.2f}%", ha='center', fontsize=11, fontweight="bold", color="black")

# Show plot with seaborn styling
sns.despine()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = "./processed_data/log/fire_eval_log/2025-03-06_22:02:36.950203/eval_test.csv"  # Change this if needed
df = pd.read_csv(file_path, sep=";")  # Assuming values are separated by semicolons

# Define metric categories
metric_labels = ["Micro Precision", "Micro Recall", "Micro F1-score",
                 "Macro Precision", "Macro Recall", "Macro F1-score"]

# Extract NER and Relation Extraction metrics
ner_scores = [
    df["ner_prec_micro"].values[0], df["ner_rec_micro"].values[0], df["ner_f1_micro"].values[0],
    df["ner_prec_macro"].values[0], df["ner_rec_macro"].values[0], df["ner_f1_macro"].values[0]
]

rel_scores = [
    df["rel_prec_micro"].values[0], df["rel_rec_micro"].values[0], df["rel_f1_micro"].values[0],
    df["rel_prec_macro"].values[0], df["rel_rec_macro"].values[0], df["rel_f1_macro"].values[0]
]

# Define bar width and X positions
bar_width = 0.4
x = range(len(metric_labels))

# Set seaborn style
sns.set_theme(style="whitegrid")

# Create figure
plt.figure(figsize=(10, 6))

# Plot bars for NER and Relation Extraction
plt.bar([i - bar_width/2 for i in x], ner_scores, width=bar_width, label="NER", color="royalblue", alpha=0.85)
plt.bar([i + bar_width/2 for i in x], rel_scores, width=bar_width, label="Relation Extraction", color="darkorange", alpha=0.85)

# Add value labels on bars
for i in range(len(x)):
    plt.text(i - bar_width/2, ner_scores[i] + 2, f"{ner_scores[i]:.2f}%", ha="center", fontsize=11, fontweight="bold", color="black")
    plt.text(i + bar_width/2, rel_scores[i] + 2, f"{rel_scores[i]:.2f}%", ha="center", fontsize=11, fontweight="bold", color="black")

# Labels and Title
plt.xticks(x, metric_labels, rotation=45, fontsize=12)
plt.ylabel("Score (%)", fontsize=12)
plt.title("NER vs. Relation Extraction Performance", fontsize=14, fontweight="bold")
plt.ylim(0, 100)

# Add grid and legend
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.legend(fontsize=12)

# Show plot
sns.despine()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Extract relation extraction metrics
rel_metrics = {
    "Metric": ["Micro Precision", "Micro Recall", "Micro F1-score",
               "Macro Precision", "Macro Recall", "Macro F1-score"],
    "Score (%)": [f"{df['rel_prec_micro'].values[0]:.2f}",
                  f"{df['rel_rec_micro'].values[0]:.2f}",
                  f"{df['rel_f1_micro'].values[0]:.2f}",
                  f"{df['rel_prec_macro'].values[0]:.2f}",
                  f"{df['rel_rec_macro'].values[0]:.2f}",
                  f"{df['rel_f1_macro'].values[0]:.2f}"]
}

# Extract NER metrics
ner_metrics = {
    "Metric": ["Micro Precision", "Micro Recall", "Micro F1-score",
               "Macro Precision", "Macro Recall", "Macro F1-score"],
    "Score (%)": [f"{df['ner_prec_micro'].values[0]:.2f}",
                  f"{df['ner_rec_micro'].values[0]:.2f}",
                  f"{df['ner_f1_micro'].values[0]:.2f}",
                  f"{df['ner_prec_macro'].values[0]:.2f}",
                  f"{df['ner_rec_macro'].values[0]:.2f}",
                  f"{df['ner_f1_macro'].values[0]:.2f}"]
}

# Convert to DataFrames for visualization
rel_df = pd.DataFrame(rel_metrics)
ner_df = pd.DataFrame(ner_metrics)

# Create figure with tight layout for better fit
fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # Smaller figure size for less white space

# Remove axes
for ax in axes:
    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)
    ax.set_frame_on(False)

# Create tables
table1 = axes[0].table(cellText=rel_df.values,
                        colLabels=rel_df.columns,
                        cellLoc="center",
                        loc="center",
                        colColours=["#2A4D69", "#4B86B4"])

table2 = axes[1].table(cellText=ner_df.values,
                        colLabels=ner_df.columns,
                        cellLoc="center",
                        loc="center",
                        colColours=["#2A4D69", "#4B86B4"])

# Adjust table styling to make them bigger
for table in [table1, table2]:
    table.auto_set_font_size(False)
    table.set_fontsize(12)  # Increase font size
    table.scale(1.5, 1.5)  # Zoom into the table by scaling it

# Set titles
axes[0].set_title("Relation Extraction Performance", fontsize=14, fontweight="bold", pad=10)
axes[1].set_title("NER Performance", fontsize=14, fontweight="bold", pad=10)

# Remove extra space
plt.tight_layout(pad=0.1)  # Minimize padding
plt.show()
```
